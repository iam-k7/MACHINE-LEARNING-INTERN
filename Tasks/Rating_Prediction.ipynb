```python
df['Cuisines'].fillna('Unknown', inplace=True)
```

```python
df.drop(['Restaurant ID', 'Restaurant Name', 'Address', 'Locality', 'Locality Verbose', 'City', 'Country Code', 'Rating color', 'Rating text', 'Votes'], axis=1, inplace=True)
```

```python
df.head()
```

```python
list1 = ['Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu', 'Price range']
```

```python
df[list1] = df[list1].replace({'Yes': 1, 'No': 0})
```

```python
df.head()
```

```python
import pandas as pd

df = pd.get_dummies(df, columns=["Currency"], drop_first=True)
```

```python
import sklearn

from sklearn.preprocessing import MultiLabelBinarizer

# 1. Convert string to list
df["Cuisines"] = df["Cuisines"].apply(lambda x: x.split(", "))

# 2. Apply MultiLabel Binarizer
mlb = MultiLabelBinarizer()
cuisine_encoded = mlb.fit_transform(df["Cuisines"])

# 3. Create DataFrame
cuisine_df = pd.DataFrame(
    cuisine_encoded,
    columns=mlb.classes_,
    index=df.index
)

# 4. Merge back
df = pd.concat([df.drop("Cuisines", axis=1), cuisine_df], axis=1)
```

```python
df.dtypes
```

```python
# Splitting features and target variable

x = df.drop('Aggregate rating', axis=1)
y = df['Aggregate rating']
```

```python
# Train-tets split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Training the model

from sklearn.tree import DecisionTreeRegressor

df = DecisionTreeRegressor( max_depth=12, random_state=42)

df.fit(x_train, y_train)
```

```python
# Evaluating the model

from sklearn.metrics import mean_squared_error, r2_score

y_pred = df.predict(x_test)

print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```
